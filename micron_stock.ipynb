```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime
%matplotlib inline
```

```python
filepath = '../../Downloads/HistoricalQuotes.csv'
```

```python
df = pd.read_csv(filepath)
```

```python
df = df.iloc[::-1]
```

```python
df['day'] = df['close'] - df['open']
```

```python
df.head()
```

```python
df['open_of_next_day'] = df.open.shift(-1)
```

```python
df['night'] = df['open_of_next_day'] - df['close']
```

```python
df['hold'] = df['open_of_next_day'] - df['open']
```

```python
df
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
x = df['date']
```

```python
y1 = df['hold']
```

```python
y2 = df['night']
```

```python
y3 = df['day']
```

```python
# red dashes, blue squares and green triangles
plt.plot(x, y1, 'r--', x, y2, 'bs', x, y3, 'g^')
plt.show()
```

```python
df['cum_day'] = df.day.cumsum()
```

```python
df['cum_night'] = df.night.cumsum()
```

```python
df['cum_hold'] = df.hold.cumsum()
```

```python
df
```

```python
x = df['date']
```

```python
y1 = df['cum_hold']
y2 = df['cum_night']
y3 = df['cum_day']
```

```python


plt.plot(x, y2, 'g', label='Night trading')
plt.plot(x, y1, 'b', label='Normal Performance')
plt.plot(x, y3, 'r', label='Day trading')
plt.legend()
plt.show()
```
